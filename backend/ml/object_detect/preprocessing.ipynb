## Crop images to from non-lateral image dataset

```python
def crop_img(img, img_new_path, crop_w_size, crop_h_size):
    

    #img = io.imread(img_originals_path + "/"+ file_name)
    h, w = img.shape[:2]
        
    img_list = []

    #print("Number of crop", h // crop_h_size, w // crop_w_size)
    for h_i in range(h // crop_h_size):
        for w_i in range(w // crop_w_size):

            img_cropped = img[h_i*crop_h_size: (h_i+1)*crop_h_size, w_i*crop_w_size: (w_i+1)*crop_w_size, :]
            img_list.append(img_cropped)
            
        
    f_split = os.path.splitext((img_new_path + "/"+ file_name))  
        
      
    print('Current File: ' + f_split[0] + '.jpg', 
          ", Size: " + str(h) + ":" + str(w),
         ", Crop: " + str(h // crop_h_size) + ":" + str( w // crop_w_size))    
    
    for img_i, img in enumerate(img_list):
        io.imsave(f_split[0] + '_'+ str(img_i) + '.jpg', img)
        
```

## Resize before cropping

```python
def resize_b4_crop_img(img, boundary):
    
    h, w = img.shape[:2]
    
    max_wh = 0
    #h_larger_bool = False
    if h >= w:
        #h_larger_bool = True
        max_wh = h
    else:
        #h_larger_bool = False
        max_wh = w
        
        
    if max_wh > boundary:
        # rescalse the image
        resize_ratio = max_wh / boundary
        
        image_resized = resize(img, (img.shape[0] // resize_ratio, img.shape[1] // resize_ratio),
                        anti_aliasing=True)
        image_resized = 255 * image_resized
        image_resized = image_resized.astype(np.uint8)
        
        #fx, plots = plt.subplots(2, 2, figsize=(20,10))        
        #plots[0][0].set_title("Images")
        #plots[0][0].imshow(image_resized)
        #print(resize_ratio, img.shape, image_resized.shape)
        
        img = image_resized
        
    return img
    
```

```python
img_originals_path = './images/non-lateral/online/tocrop'
img_new_path = './images/non-lateral/online/cropped'

# looping all the files
    
files = [f for f in listdir(img_originals_path) if isfile(join(img_originals_path, f))]
    
for file_name in files:
    img = io.imread(img_originals_path + "/"+ file_name)

    img = resize_b4_crop_img(img, 1000)
    crop_img(img, img_new_path, 128, 384)
```
